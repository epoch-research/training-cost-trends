```python
from collections import defaultdict
import pandas as pd
from pprint import pprint

from data import *
```

```python
price_df = load_price_df()
```

```python
# Filter by vendor to Microsoft Azure
avg_cud = defaultdict(dict)
vendors = ['Amazon Web Services', 'Microsoft Azure', 'Google Cloud']
price_types = ['Price per chip-hour (on-demand)', 'Price per chip-hour (1-year CUD)', 'Price per chip-hour (3-year CUD)']
for vendor in vendors:
    vendor_df = price_df[price_df['Vendor'] == vendor]
    prices_cud_1y = vendor_df.dropna(subset=['Price per chip-hour (1-year CUD)'])
    prices_cud_3y = vendor_df.dropna(subset=['Price per chip-hour (3-year CUD)'])
    cud_1y = 1 - prices_cud_1y['Price per chip-hour (1-year CUD)'] / prices_cud_1y['Price per chip-hour (on-demand)']
    cud_3y = 1 - prices_cud_3y['Price per chip-hour (3-year CUD)'] / prices_cud_3y['Price per chip-hour (on-demand)']
    print(f'{vendor} 1-year CUD: {cud_1y.mean():.2f}')
    print(f'{vendor} 3-year CUD: {cud_3y.mean():.2f}')
    avg_cud[vendor]['Price per chip-hour (1-year CUD)'] = cud_1y.mean().round(2)
    avg_cud[vendor]['Price per chip-hour (3-year CUD)'] = cud_3y.mean().round(2)
```
